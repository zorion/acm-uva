```python
def get_in(s):
    l = [i for i in s.split('\n') if i]
    return l

def next_postition(current, width, slope):
    x, y = current
    next_x = (x + slope[0]) % width
    next_y = y + slope[1]
    return next_x, next_y

def travel(l, slope=(3,1)):
    start = 0,0
    width = len(l[0])
    n = start
    while n[1] < len(l):
        yield(n)
        n = next_postition(n, width, slope)
```
