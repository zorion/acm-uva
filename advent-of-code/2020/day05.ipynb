```python
def get_row(board):
    return int(board.replace('B', '1').replace('F','0').replace('L', '').replace('R', ''), 2)

def get_column(board):
    return int(board.replace('B', '').replace('F','').replace('L', '0').replace('R', '1'), 2)

assert get_row('BFFFBBFRRR') == 70
assert get_row('FFFBBBFRRR') == 14
assert get_row('BBFFBBFRLL') == 102
assert get_column('BFFFBBFRRR') == 7
assert get_column('FFFBBBFRRR') == 7
assert get_column('BBFFBBFRLL') == 4
```

```python
def get_seat(board):
    r = get_row(board)
    c = get_column(board)
    return r * 8 + c
assert get_seat('BFFFBBFRRR') == 567
assert get_seat('FFFBBBFRRR') == 119
assert get_seat('BBFFBBFRLL') == 820
```

```python
def parse_input(s):
    return [b for b in s.split('\n') if b]
def get_highest(board_list):
    h = 0
    for b in board_list:
        aux = get_seat(b)
        if aux > h:
            h = aux
    return h
```

```python
l = parse_input(s)
print(get_highest(l))
```
