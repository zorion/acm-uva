```python
sample = """L.LL.LL.LL
LLLLLLL.LL
L.L.L..L..
LLLL.LL.LL
L.LL.LL.LL
L.LLLLL.LL
..L.L.....
LLLLLLLLLL
L.LLLLLL.L
L.LLLLL.LL
"""
```

```python
def get_board(s):
    return [list(x) for x in s.split('\n') if x]

def clone_board(board):
    return [x[:] for x in board]

def run_step(board):
    copy_board = clone_board(board)
    for i, row in enumerate(board):
        for j, col in enumerate(row):
            copy_board[i][j] = _calc_next(board, i, j)
    return copy_board

def _calc_next(board, i, j):
    if board[i][j] == '.':
        return '.'
    if board[i][j] == 'L':
        return _calc_pre_empty(board, i, j)
    if board[i][j] == '#':
        return _calc_pre_occupied(board, i, j)
    assert False, 'foo'

def _list_adjacents(board, x, y):
    neighbours = [-1, 0, 1]
    result = []
    for delta_x in neighbours:
        for delta_y in neighbours:
            if delta_x == 0 and delta_y == 0:
                # Do not add to results my own position
                continue
            a = x + delta_x
            if a < 0 or a >= len(board):
                result.append('L')
                continue
            b = y + delta_y
            if b < 0 or b >= len(board[0]):
                result.append('L')
                continue
            result.append(board[a][b])
    return result
            

def _calc_pre_empty(board, i, j):
    adjacents = _list_adjacents(board, i, j)
    if len([x for x in adjacents if x != '#']) == 8:
        return '#'
    return 'L'

def _calc_pre_occupied(board, i, j):
    adjacents = _list_adjacents(board, i, j)
    if len([x for x in adjacents if x=='#']) >= 4:
        return 'L'
    return '#'
```

```python
def run(s):
    b = get_board(s)
    old = []
    while b != old:
        old = b
        b = run_step(old)
    return len([1 for row in b for col in row if col == '#'])
            
result = run(sample)
result
```
