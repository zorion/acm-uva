```python
import re
def _testrules(a, b):
    test = apply_rules(a)
    assert test == b, '{} -> {} ({})'.format(a, test, b)
def parse_input(s):
    return s.split('\n')

def apply_rules(tile):
    touched = True
    while touched:
        aux = tile
        for match, repl in RULES:
            tile = re.sub(match, repl, tile)
            # print(aux, tile, match, repl)
        touched = aux != tile
    return tile
_testrules('esew', 'se')
_testrules('nwwswee', '')
_testrules('sesenwnenenewseeswwswswwnenewsewsw', 'seswww')

def apply_rules(tile):
    x, y = 0, 0
    i = 0
    while i < len(tile):
        # print(i, x, y, tile)
        if tile[i] == 'e':
            x -= 2
            i += 1
            continue
        if tile[i] == 'w':
            x += 2
            i += 1
            continue
        if tile[i] == 's':
            y -= 2
            i += 1
        if tile[i] == 'n':
            y += 2
            i += 1
        if tile[i] == 'e':
            x -= 1
            i += 1
            continue
        if tile[i] == 'w':
            x += 1
            i += 1
            continue
    return x, y

_testrules('esew', (-1, -2))
_testrules('nwwswee', (0,0))

def parse_tiles(tiles):
    touched = set()
    for t in tiles:
        tile = apply_rules(t)
        # print(t, tile)
        if tile in touched:
            # print('Flip', tile)
            touched.remove(tile)
        else:
            touched.add(tile)
    return touched

def problem1(s):
    tiles = parse_input(s)
    touched = parse_tiles(tiles)
    print('Problem1:', len(touched))
    return len(touched)
assert problem1(sample) == 10
assert problem1(my_in) == 377

def step(touched):
    new_floor = set()
    for tile in touched:
        if PRINT:
            print('\n   Check:', new_floor)
        if to_black(tile, touched):
            new_floor.add(tile)
        if PRINT:
            print()
        for n in get_neighbours(tile):
            if to_black(n, touched):
                new_floor.add(n)
    return new_floor

def to_black(tile, touched):
    neighs = len([n for n in get_neighbours(tile) if n in touched])
    if PRINT:
        print(tile, neighs, sorted(get_neighbours(tile)))
    if tile in touched:
        #black
        return neighs in [1,2]
    else:
        #whilte
        return neighs == 2
    return

def get_neighbours(tile):
    x, y = tile
    yield x-2,y
    yield x+2,y
    for j in [-2, 2]:
        for i in [-1, 1]:
            yield x+i, y+j

def problem2(s, many=100):
    tiles = parse_input(s)
    touched = parse_tiles(tiles)
    if PRINT:
        print('--Probl1', len(touched), sorted(touched))
    for i in range(many):
        touched = step(touched)
        if PRINT:
            print('--Step', i, 'touched:', len(touched), sorted(touched))
    result = len(touched)
    print('Problem2', result)
    return result
PRINT = False
problem2(sample, 10)
problem2(sample)
PRINT = False
problem2(my_in)
```
