```python
sample1 = """.#.
..#
###"""

my_input = """..#..#.#
##.#..#.
#....#..
.#..####
.....#..
...##...
.#.##..#
.#.#.#.#"""
```

```python
def parse_input(s):
    result = set()
    z = 0
    w = 0
    for x, line in enumerate(s.split('\n')):
        for y, col in enumerate(line):
            if col == '#':
                _key = (x, y, z, w)
                result.add(_key)
    return result, x, y
test1 = parse_input(sample1)
assert test1 == ({(0,1,0,0),(1,2,0,0),(2,0,0,0),(2,1,0,0),(2,2,0,0)}, 2, 2), test1

def get_neighbours(position):
    _ngs = [-1, 0, 1]
    for x in _ngs:
        for y in _ngs:
            for z in _ngs:
                for w in _ngs:
                    if x == 0 and y == 0 and z == 0 and w == 0:
                        continue
                    yield (position[0] + x, position[1] + y, position[2] + z, position[3] + w)
def step(status):
    minx,maxx,miny,maxy,minz,maxz, minw,maxw,board = status
    new_board = set()
    for x in range(minx-1, maxx+2):
        for y in range(miny-1, maxy+2):
            for z in range(minz-1,maxz+2):
                for w in range(minw-1, maxw+2):
                    _key = (x, y, z, w)
                    active_neighbours = len([n for n in get_neighbours(_key) if n in board])
                    if _key in board and active_neighbours in [2,3]:
                        new_board.add(_key)
                    if _key not in board and active_neighbours == 3:
                        new_board.add(_key)
    return (minx - 1, maxx + 1,
            miny - 1, maxy + 1,
            minz - 1, maxz + 1,
            minw - 1, maxw + 1,
            new_board)

def run_all(s):
    board, max_x, max_y = parse_input(s)
    status = (0, max_x, 0, max_y, 0, 0, 0, 0, board)
    for i in range(6):
        status = step(status)
    result = len(status[8])
    print('Problem2', result)
    return result
run_all(sample1)
run_all(my_input)
```

```python
def print_board(status):
    nx,xx,ny,xy,nz,xz, board = status
    for z in range(nz, xz + 1):
        print('Z=', z)
        for x in range(nx,xx +1):
            for y in range(ny,xy+1):
                if (x, y, z) in board:
                    print('#', end='')
                else:
                    print('.', end='')
            print('')
```
