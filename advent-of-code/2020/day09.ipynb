```python
def is_sumable(nums, target):
    for i, a in enumerate(nums):
        for j in range(i):
            if a + nums[j] == target:
                return False
    return True

def check_input(s, end=25):
    nums = [int(i) for i in s.split('\n') if i]
    print(len(nums))
    start = 0
    while end < len(nums):
        if is_sumable(nums[start:end], nums[end]):
            target = nums[end]
            print("not in part1", end, target)
            break
        end += 1
        start += 1
    finished = False
    for i in range(len(nums)):
        for j in range(i+2, len(nums)):
            if target == sum(nums[i:j]):
                print(i, j, nums[i:j], sum(nums[i:j]))
                start = i
                end = j
                finished = True
                break
        if finished:
            break
    minim = min(nums[start:end])
    maxim = max(nums[start:end])
    print(minim, maxim, minim+maxim)

ini = time.time()
check_input(sample, 5)
end = time.time()
print('Timing:', end-ini)
```

```python
def is_sumable(nums, target):
    for i, a in enumerate(nums):
        for j in range(i):
            if a + nums[j] == target:
                return False
    return True

def find_missing(nums, end):
    start = 0
    while end < len(nums):
        if is_sumable(nums[start:end], nums[end]):
            target = nums[end]
            print('part1', end, target)
        end += 1
        start += 1
    return target

def find_range_sum_target(nums, target):
    start = 0
    end = 2
    while True:
        assert start < end
        assert end <= len(nums)
        total = sum(nums[start:end])
        if total == target:
            break
        if start + 2 == end:
            end += 1
            continue
        if total < target:
            end += 1
            continue
        if total > target:
            start += 1
            continue
        assert False, (total, target)
    return start, end

def check_input2(s, end=25):
    nums = [int(i) for i in s.split('\n') if i]
    print('input size', len(nums))
    target = find_missing(nums, end)
    p2_start, p2_end = find_range_sum_target(nums, target)
    minim = min(nums[p2_start:p2_end])
    maxim = max(nums[p2_start:p2_end])
    print('part2', minim, maxim, minim+maxim)

ini = time.time()
check_input2(sample, 5)
end = time.time()
print('Timing', end-ini)
```
