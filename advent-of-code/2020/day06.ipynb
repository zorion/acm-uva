```python
def get_groups(s):
    return s.split('\n\n')

def get_singles(group):
    for i in group.split('\n'):
        if i:
            yield i

def parse_answer(single, answer_set):
    answer_set &= set(single)

def count_diff_answers(group):
    total = set()
    for i, s in enumerate(get_singles(group)):
        if i == 0:
            total = set(s)
        parse_answer(s, total)
    return len(total)

def get_total_answers(s):
    total = 0
    for g in get_groups(s):
        total += count_diff_answers(g)
    return total
assert get_total_answers(sample) == 6
```
