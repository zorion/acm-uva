```python
import re
def parse_tiles(s):
    result = {}
    tiles_by_num = {}
    for tile_def in s.split('\n\n'):
        header, tile = tile_def.split('\n', 1)
        tile_num = int(re.match('^Tile (\d+):', header).group(1))
        result[tile_num] = _parse_a_tile(tile)
        tiles_by_num[tile_num] = tile.split('\n')
    return result, tiles_by_num
def _parse_a_tile(tile):
    tile_lines = tile.split('\n')
    top = [v for v in tile_lines[0]]
    bottom = [v for v in tile_lines[-1][::-1]]
    left = [line[0] for line in tile_lines][::-1]
    right = [line[-1] for line in tile_lines]
    return (top, right, bottom, left)
parse_tiles(sample)[0]
```

```python
from collections import defaultdict
def compare_lines(line_a, line_b):
    return line_a == line_b or line_a == line_b[::-1]
def find_matches(tiles_map):
    result = defaultdict(list)
    for tile_a in tiles_map:
        for line_a in tiles_map[tile_a]:
            matched = False
            for tile_b in tiles_map:
                if tile_a == tile_b:
                    continue
                for i, line_b in enumerate(tiles_map[tile_b]):
                    if line_a == line_b:
                        # Invert
                        result[tile_a].append((i, tile_b, True))
                        matched = True
                    if line_a == line_b[::-1]:
                        # Do not invert
                        result[tile_a].append((i, tile_b, False))
                        matched = True
            if not matched:
                result[tile_a].append(None)
    # Check we have only one solution for each match
    assert all([4==len(v) for v in result.values()])
    return result
def find_corners(matches):
    for k, v in matches.items():
        if len([x for x in v if x]) == 2:
            yield k
find_matches(parse_tiles(sample)[0])
```

```python
def print_piece(piece):
    for line in piece:
        print(''.join([str(x) for x in line]))
    return

def rotate_piece(piece):
    to_zip = piece[::-1]
    return [list(r) for r in zip(*to_zip)]

assert rotate_piece(['12', '34']) == [['3', '1'], ['4', '2']]
assert rotate_piece([['1', '2'], ['3', '4']]) == [['3', '1'], ['4', '2']]

def invert_piece(piece):
    return [[_col for _col in _line] for _line in piece[::-1]]
assert invert_piece(['12', '34']) == [['3', '4'], ['1', '2']],invert_piece(['12', '34'])
assert invert_piece([['1', '2'], ['3', '4']]) == [['3', '4'], ['1', '2']]

def solve_puzzle(matches, p_tiles):
    result = []
    for tile_corner, tile_matches in matches.items():
        if len([v for v in tile_matches if v is None]) == 2:  # A corner
            break
    for i in range(4):
        # Find top and left
        if tile_matches[i % 4] is None and tile_matches[(i + 3) % 4] is None:
            rotation = i
            inversion = False
            break
    else:
        assert False, 'Not proper corner: %s' % tile_matches
    previous_left = tile_corner
    line, next_left, inv = solve_line(matches, tile_corner, None, p_tiles, False)
    result.append(line)
    while next_left is not None:
        line, aux, inv = solve_line(matches, next_left, previous_left, p_tiles, inv)
        previous_left = next_left
        next_left = aux
        result.append(line)
    return result

def solve_line(matches, left_tile, previous_left, p_tiles, inv):
    result = [left_tile]
    left_matches = matches[left_tile]
    next_piece, next_bottom, inv = position_left_piece(
        left_matches, previous_left, p_tiles, left_tile, inv)
    bottom_inversion = inv
    aux = left_tile
    while next_piece is not None:
        result.append(next_piece)
        next_aux = next_piece
        next_piece, inv = position_inner_piece(matches[next_piece], aux, p_tiles, next_piece, inv)
        aux = next_aux
    return result, next_bottom, bottom_inversion

def _g(piece_match):
    if piece_match is None:
        return None
    return piece_match[1]

def _rotate_piece(tile_num, p_tiles, piece_match, rotation, inversion, direction):
    # print('Rotate', tile_num, rotation, piece_match, 'invert?', inversion, 'top/left', direction)
    if piece_match is not None and piece_match[2]:
        inversion = not inversion
    if inversion:
        p_tiles[tile_num] = invert_piece(p_tiles[tile_num])
        if rotation % 2 == 0:
            rotation += 2
    rotation = (7 - rotation)
    for i in range(rotation):
        p_tiles[tile_num] = rotate_piece(p_tiles[tile_num])
    return inversion

def position_left_piece(piece_matches, top_piece, p_tiles, current, inv):
    for i in range(4):
        if piece_matches[i] is not None:
            continue
        _paux = piece_matches[(i+1)%4]
        if _paux is None and top_piece is None:
            inv = _rotate_piece(current, p_tiles, _paux, i, inv, None)
            return _g(piece_matches[(i+2)%4]), _g(piece_matches[(i+3)%4]), inv
        if top_piece is None:
            continue
        _paux = piece_matches[(i+1)%4]
        if _g(_paux) == top_piece:
            inv = _rotate_piece(current, p_tiles, _paux, i, inv, 'top')
            return _g(piece_matches[(i+2)%4]), _g(piece_matches[(i+3)%4]), inv
        _paux = piece_matches[(i+3)%4]
        if _g(_paux) == top_piece:
            inv = _rotate_piece(current, p_tiles, _paux, i, inv, 'top')
            return _g(piece_matches[(i+2)%4]), _g(piece_matches[(i+1)%4]), inv
    else:
        assert False, 'Error positioning left'
    return

def position_inner_piece(piece_matches, left_piece, p_tiles, current, inv):
    for i in range(4):
        if _g(piece_matches[i]) == left_piece:
            inv = _rotate_piece(current, p_tiles, piece_matches[i], i, inv, 'left')
            return _g(piece_matches[(i+2)%4]), inv
    else:
        assert False, 'Error positioning inner'
    return

def _rdc(piece):
    return piece[1:-1]
    return piece

def join_puzzle(tile_def, puzzle_pieces, matches):
    result = []
    prev_line = None
    for line in puzzle_pieces:
        if prev_line is None:
            pass
        for row_num in range(1, len(tile_def[line[0]])-1):
            line_data = []
            for tile in line:
                line_data += _rdc(tile_def[tile][row_num])
            result.append(line_data)
    return result

MONSTER = """                  # 
#    ##    ##    ###
 #  #  #  #  #  #   """.split('\n')
MONSTER_POSITIONS = {(i, j)
                     for i, _ in enumerate(MONSTER)
                     for j, _ in enumerate(MONSTER[i])
                     if MONSTER[i][j] == '#'}

def find_monster(puzzle, i, j):
    for a, b in MONSTER_POSITIONS:
        if i+a >= len(puzzle):
            return False
        if j+b >= len(puzzle[i+a]):
            return False
        if puzzle[i+a][j+b] == '.':
            return False
    return True

def swap_monster(puzzle, i, j):
    for a, b in MONSTER_POSITIONS:
        puzzle[a+i][b+j] = 'O'
    return

def flag_monsters(puzzle):
    for i in range(len(puzzle)):
        for j in range(len(puzzle[i])):
            if find_monster(puzzle, i, j):
                swap_monster(puzzle, i, j)
    return

def count_hashes(puzzle):
    return len([tile for line in puzzle for tile in line if tile == '#'])

def problem2(s):
    tiles, tile_def = parse_tiles(s)
    matches = find_matches(tiles)
    puzzle_pieces = solve_puzzle(matches, tile_def)
    puzzle = join_puzzle(tile_def, puzzle_pieces, matches)
    # print_piece(puzzle)
    for i in range(2):
        for j in range(4):
            puzzle = rotate_piece(puzzle)
            flag_monsters(puzzle)
            #print_piece(puzzle)
        puzzle = invert_piece(puzzle)
    result = count_hashes(puzzle)
    return result
assert problem2(sample) == 273, problem2(sample)
problem2(my_in)
```
