```python
def parse_rules(s):
    result = {}
    for line in s.split('.\n'):
        if line:
            parse_line(line, result)
    return result

def parse_line(line, result):
    key, values = line.split(' bags contain ')
    result[key] = []
    for bag in values.split(', '):
        add_bag = parse_bag(bag)
        if add_bag:
            result[key].append(add_bag)
    return

def parse_bag(bag):
    if bag == 'no other bags':
        return None
    if bag.endswith(' bag'):
        assert bag.startswith('1 ')
        return ('1', bag[2:-4])
    assert bag.endswith(' bags')
    return tuple(bag[:-5].split(' ', 1))

parse_rules(sample)
```

```python
def invert_rules(rules):
    result = defaultdict(list)
    for k, v in rules.items():
        for n, color in v:
            result[color].append(k)
    return result

invert_rules(parse_rules(sample))
```

```python
def count_in(s, color='shiny gold'):
    ir = invert_rules(parse_rules(s))
    pending = {color}
    counted = set()
    result = -1
    while pending:
        big = pending.pop()
        if big in counted:
            continue
        counted.add(big)
        result += 1
        for i in ir[big]:
            pending.add(i)
    return result

count_in(sample)
        
```

```python
def count_inner_bags(s, color='shiny gold'):
    rules = parse_rules(s)
    cache ={}
    return _count(rules, color, cache)

def _count(rules, color, cache):
    if color in cache:
        return cache[color]
    res = 0
    for n, bag in rules[color]:
        res += int(n) + int(n) * _count(rules, bag, cache)
    cache[color] = res
    return res

print(count_inner_bags(sample))
print(count_inner_bags(sample2))
count_inner_bags(my_in)
```
