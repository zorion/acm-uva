```python

def get_input(s):
    scanners = [get_scanner(x) for x in s.strip().split('\n\n')]
    return scanners


def get_scanner(raw):
    result = set()
    for point in raw.split('\n')[1:]:
        t = tuple([int(x) for x in point.split(',')])
        assert len(t) == 3, t
        result.add(t)
    return result


print([len(x) for x in get_input(sample_in)])
print([len(x) for x in get_input(my_in)])
```

```python
def prob1(s):
    scanners = get_input(s)
    corrected = {0: (scanners[0], (0,0,0))}
    placed = {0}
    while placed:
        item_num = placed.pop()
        beacons = corrected[item_num][0]
        for i, messy_scanner in enumerate(scanners):
            if i in corrected:
                continue
            scan_ok, scan_oriented, position = get_matching(beacons, messy_scanner)
            if scan_ok:
                corrected[i] = (scan_oriented, position)
                placed.add(i)
    assert len(scanners) == len(
        corrected), f'{len(corrected)} != {len(scanners)}'
    result = count_beacons(corrected)
    return result


def count_beacons(corrected):
    res = set()
    for beacons in corrected.values():
        res |= beacons[0]
    return len(res)


def get_matching(beacons, messy_beacons):
    for intent in get_orientations(messy_beacons):
        distances = get_distances(beacons, intent)
        options = [k for k, v in distances.items() if v >= 12]
        if options:
            assert len(options) == 1
            return True, re_position(intent, options[0]), options[0]
    return False, None, None


def get_distances(beacons, messy_beacons):
    result = {}
    for beacon in beacons:
        for mb in messy_beacons:
            pos = get_vector(beacon, mb)
            if pos not in result:
                result[pos] = 0
            result[pos] += 1
    return result


def get_vector(x, y):
    return (x[0]-y[0], x[1]-y[1], x[2]-y[2])


def re_position(messy_beacons, vector):
    return {tuple([m[i]+vector[i] for i in range(3)])for m in messy_beacons}


def get_orientations(messy_beacons):
    for i in [-1, 1]:
        for j in [-1, 1]:
            for k in [-1, 1]:
                for permut in [(0, 1, 2), (0, 2, 1),
                               (1, 0, 2), (1, 2, 0),
                               (2, 0, 1), (2, 1, 0),
                               ]:
                    yield {tuple([x*beacon[y] for x, y in zip((i, j, k), permut)])
                           for beacon in messy_beacons}
    return

test_orient = list(get_orientations({(0,1,2), (3,4,5)}))
assert {(0, -1, -2), (3, -4, -5)} in test_orient
assert {(0, -1, -2), (-3, -4, -5)} in test_orient
assert {(2, 1, 0), (5, 4, -3)} in test_orient
assert len(test_orient) == 48, len(test_orient)

test1 = prob1(sample_in)
assert test1 == 79, test1
prob1(my_in)
```
