```python
sample_in = """7,4,9,5,11,17,23,2,0,14,21,24,10,16,13,6,15,25,12,22,18,20,8,19,3,26,1

22 13 17 11  0
 8  2 23  4 24
21  9 14 16  7
 6 10  3 18  5
 1 12 20 15 19

 3 15  0  2 22
 9 18 13 17  5
19  8  7 25 23
20 11 10 24  4
14 21 16 12  6

14 21 17 24  4
10 16 15  9 19
18  8 23 26 20
22 11 13  6  5
 2  0 12  3  7
"""
N = 5
class Board:
    def __init__(self, lines):
        self.lines = lines
        self.unmarked = {_i for _l in self.lines[:N] for _i in _l}
        assert len(self.unmarked) == N * N
    def __repr__(self):
        return '[' + '], ['.join([' '.join(map(str,_l)) for _l in self.lines[:N]]) + ']'
def get_board(d):
    lines = d.split('\n')
    rows = [[int(_l.strip()) for _l in line.split(' ') if _l] for line in lines]
    assert len(rows) == N
    cols = []
    for i in range(N):
        cols.append([_r[i] for _r in rows])
    return Board(rows+cols)
def get_input(s):
    data = s.strip().split('\n\n')
    winners = [int(_d) for _d in data[0].strip().split(',')]
    boards = [get_board(_d) for _d in data[1:]]
    return winners, boards
print(get_input(sample_in))
```

```python
def check_board(nums, board):
    winner_line = None
    num_winners = 0
    for line in board.lines:
        if set(line) < set(nums):
            winner_line = board.unmarked
            num_winners += 1
            print(f'Winner #{num_winners}:', winner_line)
    return winner_line

def score_line(line):
    return sum(line)

def prob1(s):
    winners, boards = get_input(s)
    drawn = []
    for ball in winners:
        drawn.append(ball)
        for board in boards:
            if ball in board.unmarked:
                board.unmarked.remove(ball)
                res = check_board(drawn, board)
                if res is not None:
                    # board winner with res being the winner line and ball the last ball
                    print(ball, score_line(res))
                    return ball * score_line(res)
assert prob1(sample_in) == 4512
```
