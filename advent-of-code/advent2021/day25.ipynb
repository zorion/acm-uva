```python
sample_in = """
v...>>.vv>
.vv>>.vv..
>>.>v>...v
>>v>>.>.v.
v>v.vv.v..
>.>>..v...
.vv..>.>v.
v.v..>>v.v
....v..v.>"""

def get_input(s):
    result = {}
    for i, line in enumerate(s.strip().split('\n')):
        for j, cell in enumerate(line.strip()):
            if cell == '.':
                pass
            elif cell == '>':
                result[(i, j)] = True
            elif cell == 'v':
                result[(i, j)] = False
            else:
                assert False, (i, j, cell)
    return result, i+1, j+1

def draw_board(board, rows, cols):
    for i in range(rows):
        for j in range(cols):
            if (i, j) not in board:
                print('.', end='')
            elif board[(i, j)]:
                print('>', end='')
            else:
                print('v', end='')
        print()
print(get_input(sample_in))
```

```python
def run_step(board, rows, cols):
    is_moved = False
    east_movers = {k for k,v in board.items() if v}
    south_movers = {k for k,v in board.items() if not v}

    new_board = {}
    removers = set()
    for em in east_movers:
        i, j = em
        next_j = (j+1) %cols
        if (i, next_j) in board:
            new_board[(i, j)] = True
        else:
            new_board[(i, next_j)] = True
            board[(i, next_j)] = True
            removers.add(em)
            is_moved = True
    for em in removers:
        del board[em]

    removers = set()
    for sm in south_movers:
        i, j = sm
        next_i = (i+1) %rows
        if (next_i, j) in board:
            new_board[(i, j)] = False
        else:
            new_board[(next_i, j)] = False
            is_moved = True
    return is_moved, new_board

def prob1(s):
    board, rows, cols = get_input(s)
    unfinished = True
    step_num = 0
    while unfinished:
        step_num += 1
        unfinished, board = run_step(board, rows, cols)
    return step_num

assert prob1(sample_in) == 58

```
