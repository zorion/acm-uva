```python
def get_input(s):
    lines = map(int, s.strip().split('\n'))
    return lines

def get_increases_num(depths):
    result = 0
    prev = next(depths)
    for l in depths:
        if l > prev:
            result += 1
        prev = l
    return result

def realign_not_iterator(depths):
    for i in range(len(depths)-2):
        measurement = depths[i] + depths[i+2] + depths[i+1]
        yield measurement

def realign(depths):
    a = next(depths)
    b = next(depths)
    c = next(depths)
    for x in depths:
        yield a + b + c
        a = b
        b = c
        c = x
    yield a + b + c

def prob1(s):
    return get_increases_num(get_input(s))

def prob2(s):
    # measurements = realign_not_iterator(list(get_input(s)))
    measurements = realign(get_input(s))
    return get_increases_num(measurements)
    

print(prob1(sample1))
print(prob2(sample1))
```
