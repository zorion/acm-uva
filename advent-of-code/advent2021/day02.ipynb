```python
sample_in = """forward 5
down 5
forward 8
up 3
down 8
forward 2"""

def get_input_list(s):
    return s.strip().split('\n')
get_input_list(sample_in)
```

```python
def run(instruction, position):
    move, delta = instruction.split(' ')
    delta = int(delta)
    if move == 'forward':
        return position[0] + delta, position[1]
    if move == 'down':
        return position[0], position[1] + delta
    if move == 'up':
        return position[0], position[1] - delta

def prob1(s):
    instructions = get_input_list(s)
    position = (0, 0)
    for i in instructions:
        position = run(i, position)
    print(position)
    return position[0] * position[1]
prob1(sample_in)
```

```python
def run2(instruction, position):
    move, delta = instruction.split(' ')
    delta = int(delta)
    if move == 'forward':
        return position[0] + delta, position[1] + (position[2] * delta), position[2]
    if move == 'down':
        return position[0], position[1], position[2] + delta
    if move == 'up':
        return position[0], position[1], position[2] - delta

def prob2(s):
    instructions = get_input_list(s)
    position = (0, 0, 0)
    for i in instructions:
        position = run2(i, position)
    print(position)
    return position[0] * position[1]
prob2(sample_in)
```
