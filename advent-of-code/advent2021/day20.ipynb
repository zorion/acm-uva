```python
sample_in = """
..#.#..#####.#.#.#.###.##.....###.##.#..###.####..#####..#....#..#..##..###..######.###...####..#..#####..##..#.#####...##.#.#..#.##..#.#......#.###.######.###.####...#.##.##..#..#..#####.....#.#....###..#.##......#.....#..#..#..##..#...##.######.####.####.#.#...#.......#..#.#.#...####.##.#......#..#...##.#.##..#...##.#.##..###.#......#.#.......#.#.#.####.###.##...#.....####.#..#..#.##.#....##..#.####....##...##..#...#......#.#.......#.......##..####..#...#.#.#...##..#.#..###..#####........#..####......#..#

#..#.
#....
##..#
..#..
..###
"""


def get_input(s):
    algo, grid = s.strip().split('\n\n')
    grid_dict = set()
    for i, line in enumerate(grid.split('\n')):
        for j, v in enumerate(line):
            if v == '.':
                pass
            elif v == '#':
                grid_dict.add((i, j))
            else:
                assert False, v
    return algo, grid_dict


print(get_input(sample_in))
```

```python
def prob1(s, debug=False):
    algo, grid = get_input(s)
    for i in range(2):
        draw(grid, debug)
        grid = apply_step(algo, grid, step=i)
        draw(grid, debug)
    return len(grid)
def apply_step(algo, grid, step):
    result = set()
    for lighted in grid:
        if algo[0] == '.':
            for dot in get_neighbours(lighted):
                if is_light(dot, grid, algo):
                    result.add(dot)
        else:
            min_x = min([algo])
    return result
def get_neighbours(position):
    x, y = position
    return [(x+i, y+j) for i in [-1, 0 , 1] for j in [-1, 0, 1]]
def is_light(position, grid, algo):
    bin_pos = ''
    x, y = position
    for i in [-1, 0, 1]:
        for j in [-1, 0, 1]:
            if (x+i, y+j) in grid:
                bin_pos += '1'
            else:
                bin_pos += '0'
    data_pos = int(bin_pos, 2)
    return algo[data_pos] == '#'

def draw(grid, debug):
    if not debug:
        return
    min_x = min([x[0] for x in grid])-1
    max_x = max([x[0] for x in grid])+2
    min_y = min([x[1] for x in grid])-1
    max_y = max([x[1] for x in grid])+2
    result = ''
    for x in range(min_x, max_x):
        for y in range(min_y, max_y):
            if (x, y) in grid:
                result += '#'
            else:
                result += '.'
        result += f'{x}\n'
    print(result)
    return

test1 = prob1(sample_in, debug=True)
assert test1 == 35, test1
print('ok')
```

```python
def get_input(s):
    algo, grid = s.strip().split('\n\n')
    return algo, grid.split('\n')
print(get_input(sample_in))

def prob1(s):
    outside = '.'
    algo, grid = get_input(s)
    for _ in range(2):
        #print('grid', grid)
        grid, outside = apply_step(algo, grid, outside)
    return count_light(grid)
def apply_step(algo, grid, outside):
    result = []
    if algo[0] == '#':
        if outside == '.':
            new_outside = '#'
        if outside == '#':
            new_outside = '.'
    else:
        new_outside = outside
    preline = ''
    for y in range(-1, len(grid[0])+1):
        preline += get_light((-1, y), grid, outside, algo)
    result.append(preline)
    for x, line in enumerate(grid):
        thisline = get_light((x, -1), grid, outside, algo)
        for y, _ in enumerate(line):
            thisline += get_light((x, y), grid, outside, algo)
        thisline += get_light((x, y + 1), grid, outside, algo)
        result.append(thisline)
    postline = ''
    for y in range(-1, len(grid[0])+1):
        postline += get_light((x+1, y), grid, outside, algo)
    result.append(postline)
    return result, new_outside
def get_light(position, grid, outside, algo):
    x, y = position
    bingpos = ''
    for i in [-1, 0, 1]:
        for j in [-1, 0, 1]:
            if x+i < 0 or y+j < 0 or x+i >= len(grid) or y+j >= len(grid[0]):
                val = outside
            else:
                try:
                    val = grid[x+i][y+j]
                except:
                    print('error in ', x+i, y+j)
                    raise
            assert val in '#.'
            bingpos += '1' if val == '#' else '0'
    posdata = int(bingpos, 2)
    return algo[posdata]
def count_light(grid):
    result = 0
    for line in grid:
        for val in line:
            if val == '#':
                result += 1
            elif val == '.':
                pass
            else:
                assert False, val
    return result
    
test1 = prob1(sample_in)
assert test1 == 35, test1
print('ok')
```
