```python
def prob1(s):
    dots, fold_instructions = get_input(s)
    for fold_inst in fold_instructions:
        fold(dots, fold_inst)
        break
    return len(dots)
def fold(dots, instr):
    axis, qty = instr
    if axis == 'x':
        foldx(dots, qty)
    elif axis == 'y':
        foldy(dots, qty)
    else:
        assert f'Axis incorrect: {axis} (expected x or y)'
    return
def foldx(dots, qty):
    dotkeys = list(dots.keys())
    for dot in dotkeys:
        x, y = dot
        if x > qty:
            del dots[(x, y)]
            dots[(2 * qty - x, y)] = '#'
        if x == qty:
            del dots[(x, y)]
    return
def foldy(dots, qty):
    dotkeys = list(dots.keys())
    for dot in dotkeys:
        x, y = dot
        if y > qty:
            del dots[(x, y)]
            dots[(x, 2 * qty - y)] = '#'
        if y == qty:
            del dots[(x, y)]
    return
testsample = prob1(sample_in)
assert testsample == 17, testsample
assert prob1(my_in) == 607
```
