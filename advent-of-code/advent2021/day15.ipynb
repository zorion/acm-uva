```python
def prob1(s):
    board = get_input(s)
    assert len(board) == len(board[0])
    cache = {}
    start = (0, 0)
    end = (len(board)-1, len(board)-1)
    precalculate_cache(board, end, cache)
    return cache[start]
import time
def precalculate_cache(board, end, cache):
    start_time = time.time()
    inf = sum([sum(line) for line in board]) + 1
    print(f'Board of size {len(board)} with total value {inf - 1}')
    for i, line in enumerate(board):
        for j, _ in enumerate(line):
            cache[(i, j)] = inf
    cache[end] = 0
    pending_nodes = {end}
    while pending_nodes:
        node = pending_nodes.pop()
        nodeval = board[node[0]][node[1]]
        for n in get_neighbours(board, node):
            if cache[n] > nodeval + cache[node]:
                pending_nodes.add(n)
                cache[n] = nodeval + cache[node]
    end_time = time.time()
    print(f'Precalculated result={cache[(0,0)]} in {end_time - start_time} s')
    return

def get_neighbours(board, x):
    result = []
    for i in [-1]:
        if x[0] + i < len(board) and x[0] + i >= 0:
            result.append((x[0] + i, x[1]))
        if x[1] + i < len(board) and x[1] + i >= 0:
            result.append((x[0], x[1] + i))
    return result

test1 = prob1(sample_in)
assert test1 == 40, test1
assert prob1(my_in) == 748
```

```python
def prob2(s):
    smallboard = get_input(s)
    board = redefine_board(smallboard)
    assert len(board) == len(board[0])
    cache = {}
    start = (0, 0)
    end = (len(board)-1, len(board)-1)
    precalculate_cache(board, end, cache)
    return cache[start]

def redefine_board(board):
    result = []
    for multiplier in range(5):
        for row in board:
            longline = get_long_line(row, multiplier)
            result.append(longline)
    return result
def get_long_line(small_line, line_add):
    long_line = []
    for i in range(5):
        long_line += add_to_line(small_line, line_add + i)
    return long_line
def add_to_line(line, incr):
    return [((x + incr - 1) % 9) + 1 for x in line]
    
test2 = prob2(sample_in)
assert test2 == 315, test2
prob2(my_in)
```
