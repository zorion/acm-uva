```python
def prob1(s):
    instructions = get_input(s)
    cubes = set()
    for instr in instructions:
        apply_instr(cubes, instr, part=1)
        # print(f'applied {instr}: {len(cubes)}')
    # print(cubes)
    print(f'Result: {len(cubes)}')
    return len(cubes)


def apply_instr(cubes, instruction, part=1):
    code, x1, x2, y1, y2, z1, z2 = instruction
    for new_cube in get_cubes(x1, x2, y1, y2, z1, z2, part):
        if code == 'on':
            cubes.add(new_cube)
        elif code == 'off':
            if new_cube in cubes:
                cubes.remove(new_cube)
        else:
            assert False, f' expected on or off but received {code}'
    return


def get_cubes(x1, x2, y1, y2, z1, z2, part):
    result = []
    if part == 1:
        x1, x2 = clip(x1, x2, -50, 50)
        y1, y2 = clip(y1, y2, -50, 50)
        z1, z2 = clip(z1, z2, -50, 50)
    for x in range(x1, x2+1):
        for y in range(y1, y2+1):
            for z in range(z1, z2+1):
                result.append((x, y, z))
    # print(x1, x2, y1, y2, z1, z2, result)
    return result


def clip(start, end, xmin, xmax):
    if start > xmax or end < xmin:
        return 0, -1
    return min(xmax, max(start, xmin)), min(xmax, max(end, xmin))


test1 = prob1(sample_in1)
assert test1 == 39, test1
test2 = prob1(sample_in2)
assert test2 == 590784, test2
assert prob1(my_in) == 547648
```

```python
import tqdm

def prob2(s):
    instructions = get_input(s)
    return run(instructions)

def prob1(s):
    instructions = get_input(s)
    instructions_prob1 = gransform(instructions)
    return run(instructions_prob1)

def gransform(instructions):
    result = []
    for ins in instructions:
        add_res = True
        for i in [1, 3, 5]:
            if ins[i] > 50:
                add_res = False
        for i in [2, 4, 6]:
            if ins[i] < -50:
                add_res = False
        if add_res:
            result.append((ins[0],
            max(-50, ins[1]), min(50, ins[2]),
            max(-50, ins[3]), min(50, ins[4]),
            max(-50, ins[5]), min(50, ins[6])))
    return result
                

def run(instructions):
    paralepipedes = set()  # Non intersecting
    for instr in tqdm.tqdm(instructions):
        mode = instr[0]
        ppp = tuple(instr[1:])
        if mode == 'on':
            add_ppp(paralepipedes, ppp)
        elif mode == 'off':
            remove_ppp(paralepipedes, ppp)
        else:
            assert False, mode
    result = get_size(paralepipedes)
    print(f'Result: {result}')
    return result


def get_intersection(ppp1, ppp2):
    return tuple([max(ppp1[i], ppp2[i]) if i % 2 == 0 else min(ppp1[i], ppp2[i]) for i in range(6)])
assert get_intersection((1,3,1,3,1,3), (0,2,2,4,5,6)) == (1,2,2,3,5,3)
assert get_intersection((1,3,1,3,1,3), (0,2,2,4,0,1)) == (1,2,2,3,1,1)

def get_size(ppps):
    #print(ppps)
    result = 0
    for ppp in ppps:
        aux = get_ppp_size(ppp)
        # print(ppp, aux, result)
        result += aux
    return result


def get_ppp_size(ppp):
    result = 1
    for i in range(3):
        if ppp[2*i+1] < ppp[2*i]:
            return 0
        result *= (ppp[2*i + 1] - ppp[2*i] + 1)
    return result

assert get_ppp_size((1,1,1,1,1,1)) == 1
assert get_ppp_size((1,2,1,1,1,1)) == 2
assert get_ppp_size((1,2,-1,1,1,1)) == 6
assert get_ppp_size((1,2,1,2,1,2)) == 8
assert get_ppp_size((1,2,1,-2,1,2)) == 0


def add_ppp(paralelepipedes, ppp):
    new_pars = {ppp}
    for par in paralelepipedes:
        to_remove = set()
        for new_par in new_pars:
            inter = get_intersection(par, new_par)
            to_remove.add(inter)
        for rem_par in to_remove:
            remove_ppp(new_pars, rem_par)
    paralelepipedes |= new_pars
    return new_pars

def remove_ppp(paralelepipedes, ppp):
    if get_ppp_size(ppp) < 1:
        return
    remove_ppps = set()
    add_ppps = set()
    for par in paralelepipedes:
        inter = get_intersection(par, ppp)
        if get_ppp_size(inter) > 0:
            remove_ppps.add(par)
            add_ppps |= add_fragments(par, inter)
    for par in remove_ppps:
        paralelepipedes.remove(par)
    for par in add_ppps:
        paralelepipedes.add(par)

def add_fragments_long_version(ppp, intersection):
    # There are 27 potential fragments, one of them being the intersection
    result = set()
    for section_x in [-1, 0, 1]:
        for section_y in [-1, 0, 1]:
            for section_z in [-1, 0, 1]:
                if section_x == -1:
                    x1 = ppp[0]
                    x2 = intersection[0] - 1
                elif section_x == 0:
                    x1 = max(intersection[0], ppp[0])
                    x2 = min(intersection[1], ppp[1])
                else:
                    x1 = intersection[1] + 1
                    x2 = ppp[1]
                if section_y == -1:
                    y1 = ppp[2]
                    y2 = intersection[2] - 1
                elif section_y == 0:
                    y1 = max(intersection[2], ppp[2])
                    y2 = min(intersection[3], ppp[3])
                else:
                    y1 = intersection[3] + 1
                    y2 = ppp[3]
                if section_z == -1:
                    z1 = ppp[4]
                    z2 = intersection[4] - 1
                elif section_z == 0:
                    z1 = max(intersection[4], ppp[4])
                    z2 = min(intersection[5], ppp[5])
                else:
                    z1 = intersection[5] + 1
                    z2 = ppp[5]
                if (section_x, section_y, section_z) == (0,0,0):
                    continue
                fragment = (x1,x2,y1,y2,z1,z2)
                if get_ppp_size(fragment) > 0:
                    result.add(fragment)
    return result
                    
def add_fragments(ppp, intersection):
    # There are 27 potential fragments, one of them being the intersection
    result = set()
    v = [0, 0, 0]
    for v[0] in [-1, 0, 1]:
        for v[1] in [-1, 0, 1]:
            for v[2] in [-1, 0, 1]:
                if v == [0,0,0]:
                    continue  # This one is the intersection (exclude)
                x = [0, 0, 0, 0, 0, 0]
                for i in range(3):
                    if v[i] == -1:
                        x[2*i] = ppp[2*i]
                        x[2*i+1] = intersection[2*i] - 1
                    elif v[i] == 0:
                        x[2*i] = max(intersection[2*i], ppp[2*i])
                        x[2*i+1] = min(intersection[2*i+1], ppp[2*i+1])
                    else:
                        x[2*i] = intersection[2*i+1] + 1
                        x[2*i+1] = ppp[2*i+1]
                if get_ppp_size(tuple(x)) > 0:
                    result.add(tuple(x))
    return result
                    
test1 = prob1(sample_in1)
assert test1 == 39, test1
test2 = prob1(sample_in2)
assert test2 == 590784, test2
assert prob1(my_in) == 547648

test1 = prob2(sample_in1)
assert test1 == 39, test1
test1 = prob2(sample_in3)
assert test1 == 2758514936282235, test1
prob2(my_in)
```
