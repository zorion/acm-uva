```python
sample_in = """
[({(<(())[]>[[{[]{<()<>>
[(()[<>])]({[<{<<[]>>(
{([(<{}[<>[]}>{[]{[(<()>
(((({<>}<{<{<>}{[]{[]{}
[[<[([]))<([[{}[[()]]]
[{[{({}]{}}([{[{{{}}([]
{<[[]]>}<{[{[{[]{()[[[]
[<(<(<(<{}))><([]([]()
<{([([[(<>()){}]>(<<{{
<{([{{}}[<[[[<>{}]]]>[]]
"""

def get_input(s):
    return s.strip().split('\n')

SIGNS = (
    ('(', ')', 3),
    ('[', ']', 57),
    ('{', '}', 1197),
    ('<', '>', 25137),
)
SIGN_D = {s[1]: (s[0], s[2]) for s in SIGNS}
SIGN_P = {s[0]: i for i, s in enumerate(SIGNS, 1)}

get_input(sample_in)
```

```python
def prob1(s):
    result = 0
    lines = get_input(s)
    for line in lines:
        result += parse_line(line)
    print(f'prob1 with {len(s)}chars and {len(lines)}lines with result {result}')
    return result
def parse_line(line):
    global SIGN_D
    stack = []
    for ch in line:
        if ch in SIGN_D:
            if stack and stack[-1] == SIGN_D[ch][0]:
                stack.pop()
            else:
                return SIGN_D[ch][1]
        else:
            stack.append(ch)
    return 0
    
assert prob1(sample_in) == 26397, prob1(sample_in)
assert prob1(my_in) == 343863
```

```python
def prob2(s):
    lines = get_input(s)
    results = []
    for line in lines:
        aux = parse_line(line)
        if aux:
            results.append(aux)
    sres = sorted(results)
    print(sres)
    middle = (len(results)) // 2
    return sres[middle]
def parse_line(line):
    pending = get_completing(line)
    return get_score(pending)
def get_completing(line):
    global SIGN_D
    stack = []
    for ch in line:
        if ch in SIGN_D:
            if stack and stack[-1] == SIGN_D[ch][0]:
                stack.pop()
            else:
                return []
        else:
            stack.append(ch)
    return stack[::-1]
def get_score(pending):
    result = 0
    for ch in pending:
        result = result * 5 + SIGN_P[ch]
    return result
assert prob2(sample_in) == 288957, prob2(sample_in)
prob2(my_in)
```
